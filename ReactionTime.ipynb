# Read positions from desktop for click

```python
import pyautogui
import time
import keyboard  # pip install keyboard

def print_mouse_positions():
    print("Move the mouse to a position and press 'p' to print the coordinates.")
    print("Press 'q' to quit.")

    try:
        while True:
            if keyboard.is_pressed('p'):  # Press 'p' to print the current mouse position
                x, y = pyautogui.position()
                print(f"Position: {x}, {y}")
                time.sleep(0.5)  # Delay to prevent multiple prints from single press
            elif keyboard.is_pressed('q'):  # Press 'q' to quit the loop
                print("Exiting...")
                break
            time.sleep(0.1)  # Short delay to reduce CPU usage
    except KeyboardInterrupt:
        print("Program exited.")

if __name__ == "__main__":
    print_mouse_positions()

```

# Select Target Automatically from AI Debug in BeamNG

```python
import time
import subprocess
import pyautogui
from datetime import datetime
from pynput import keyboard

# Path to the BeamNG.drive executable
beamng_path = r"C:\Program Files (x86)\Steam\steamapps\common\BeamNG.drive\Bin64\BeamNG.drive.x64.exe"

# Launch BeamNG.drive with Vulkan
subprocess.Popen([beamng_path, "-force-vulkan"])


# Wait for the BeamNG scene to load
time.sleep(30)

# Press and hold the Alt key
pyautogui.keyDown('alt')

# Press the U key
pyautogui.press('u')

# Release the Alt key
pyautogui.keyUp('alt')

# Get the BeamNG window
beamng_window = pyautogui.getWindowsWithTitle("BeamNG")[0]

# Activate the BeamNG window
beamng_window.activate()

# Wait for a short time to ensure the window is in focus
time.sleep(30)

# Press and hold the Alt key
pyautogui.keyDown('alt')

# Press the U key
pyautogui.press('u')

# Release the Alt key
pyautogui.keyUp('alt')

time.sleep(2)

# Click on the position (232, 1025) to select the "AI Debug" dropdown menu
pyautogui.click(232, 1025)
pyautogui.click(232, 1025)

# Move the cursor to the position (125, 869)
pyautogui.moveTo(125, 869)

# Record the time before clicking
click_time = datetime.now()

time.sleep(1)  # Pause to see where the cursor goes
pyautogui.click()

# Press and hold the Alt key
pyautogui.keyDown('alt')

# Press the U key
pyautogui.press('u')

# Release the Alt key
pyautogui.keyUp('alt')
# Wait for the user to press the space key
print("Press the space key when ready...")
space_key_pressed = False

def on_press(key):
    global space_key_pressed, space_key_time
    if key == keyboard.Key.space:
        space_key_time = datetime.now()
        space_key_pressed = True
        return False

listener = keyboard.Listener(on_press=on_press)
listener.start()

while not space_key_pressed:
    time.sleep(0.1)

listener.stop()

# Calculate the time difference
time_difference = space_key_time - click_time
print(f"Time difference between mouse click and space key press: {time_difference}")
```
